```python
# Primeiro argumento é o arquivo a ser lido
# O "delimiter=';'" é o delimitador usado para separação no arquivo csv
# O "encoding='ISO-8859-1'" é necessário para que o pandas consiga ler as palavras com acentuação
# O "skiprows=1" é necessario para o pandas pular a primeira linha do arquivo que não faz parte do DataFrame

df_inicial = pd.read_csv('AerodromosPublicos.csv', delimiter=';', encoding='ISO-8859-1', skiprows=1)
```

```python
df_inicial
```

```python
# Criando uma lista de listas contendo apenas as colunas desejadas

list_cols = []
cols = ['Código OACI','UF','Município Servido','LATGEOPOINT','LONGEOPOINT']

for column in df_inicial.columns:
    li = df_inicial[column].tolist()
    if column in cols:
      list_cols.append(li)
```

```python
print(list_cols)
```

```python
# Criando um dicicionário com a lista das colunas desejadas e convertando para DataFrame

dicionario = {'oaci':list_cols[0],'uf':list_cols[1],'municipio':list_cols[2],'longitude':list_cols[3],'latitude':list_cols[4]}
df = pd.DataFrame(dicionario)
```

```python
df
```

```python
# Exportando o DataFrame em formato .csv

df.to_csv('df_lista_oaci.csv', index=False)
```
